```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

#Loading dataset

df = pd.read_csv("../data/archive/creditcard.csv")
```

```python
#removing duplicates from the data

df.drop_duplicates()
```

```python
#Seperating Features-

X = df.drop("Class",axis=1) #dropping class column entirely

y = df["Class"]
```

```python
#Splitting the testing and training data
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y)
```

```python
#Scale the data so larger values don't dominate.
# to improve model performance and avoid data leakage.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
# What this includes:
## Objective
Prepare the raw credit card transaction dataset for machine learning by cleaning and transforming it into a format suitable for model training.

---

## What We Did

### 1. Removed Duplicate Records
- Checked for duplicate transactions.
- Removed duplicate rows to prevent bias during model training.

### 2. Created Features (X) and Target (y)
- X (Features):** All input variables except `Class`.
- y (Target):** The `Class` column, where:
  - `0` = Legitimate Transaction
  - `1` = Fraudulent Transaction

> We excluded the `Class` column from X because the model should predict it. Including it as an input would cause **data leakage**.

### 3. Split the Dataset
- Divided the dataset into:
  - 80% Training Data
  - 20% Testing Data
- The training set is used to learn patterns.
- The test set is used to evaluate the model on unseen data.

### 4. Applied Feature Scaling
- Used StandardScaler to standardize the features.
- Scaling prevents features with large numerical values (such as `Amount`) from dominating the learning process.

### 5. Learned the Difference Between `fit()` and `transform()`
- fit() → Learns statistics (mean and standard deviation).
- transform() → Applies the learned transformation.
- fit_transform() → Performs both operations together.

> We used `fit_transform()` only on the training data and `transform()` on the test data to avoid **data leakage**.

---

## Key Concepts Learned
- DataFrame
- Features (X)
- Target (y)
- Train-Test Split
- Feature Scaling
- StandardScaler
- fit()
- transform()
- fit_transform()
- Data Leakage

---

## Outcome

The dataset is now clean, split, and standardized. It is fully prepared for training a machine learning classification model in the next stage of the project.
```
